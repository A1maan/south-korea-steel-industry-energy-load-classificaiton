# **4. k-NN Models**

## **4.1.Data Preparation**

### **4.1.1.Scaling the Features**

```python
X = df.iloc[:, :-1].values # Input Features
y = df.iloc[:, -1].values # Output column

# Generating a train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initializing a Standard scaler object
scaler = StandardScaler()
scaler.fit(X_train) # Training the scaler object on the training dataset only

# Scaling the feature values
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

## **4.2. k-NN using Euclidean Distance**

```python
# k-values range
k_values = range(1, 51)

# Creating a KFold object for cross validation
kf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 42)

# Storing the cross-validation mean accuracy in a dictionary
mean_scores_euc = {}

# Storing the mean error for each k-value
euc_error = []

for k in k_values:

    # Creating a k-NN model obejct
    knn_euc = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')

    scores_euc = cross_val_score(knn_euc, X_train, y_train, cv=kf, scoring='accuracy')

    mean_scores_euc[k] = [scores_euc.mean(), scores_euc.std()]
    print(f"k = {k}, Cross-validation mean accuracy: {scores_euc.mean()}")

    # Fitting the model to the training data
    knn_euc.fit(X_train, y_train)

    #y_pred = cross_val_predict(knn_euc, X_train, y_train, cv=kf)

    # Making predictions using the test data
    pred_i_euc = knn_euc.predict(X_test)

    # Printing the classification report and the confusion matrix
    print(classification_report(y_test, pred_i_euc))
    print(confusion_matrix(y_test, pred_i_euc))
    print("=" * 50)

    euc_error.append(np.mean(pred_i_euc != y_test))
```

```python
for k, accuracy in mean_scores_euc.items():
  print(f"k = {k}, Cross-validation mean accuracy: {accuracy[0]*100}% with a standard deviation of {accuracy[1] * 100}")
```

### **4.2.1. Error v.s K-values plot (Euclidean Distance)**

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 51), euc_error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. k Value (Distance Metric: Euclidean)')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

## **4.3. k-NN using Manhattan Distance**

```python
# k-values range
k_values = range(1, 51)

# Creating a KFold object for cross-validation
kf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 42)

# Storing the cross-validation mean accuracy in a dictionary
mean_scores_manh = {}

# Storing the mean error for each k-value
manh_error = []

for k in k_values:

    # Creating a k-NN model obejct
    knn_manh = KNeighborsClassifier(n_neighbors=k, metric = 'manhattan')

    scores_manh = cross_val_score(knn_manh, X_train, y_train, cv=kf, scoring='accuracy')

    mean_scores_manh[k] = [scores_manh.mean(), scores_manh.std()]
    print(f"k = {k}, Cross-validation mean accuracy: {scores_manh.mean()}")

    # Fitting the model to the training data
    knn_manh.fit(X_train, y_train)

    #y_pred = cross_val_predict(knn_manh, X_train, y_train, cv=kf)

    # Making predictions using the test data
    pred_i_manh = knn_manh.predict(X_test)

    # Printing the classification report and the confusion matrix
    print(classification_report(y_test, pred_i_manh))
    print(confusion_matrix(y_test, pred_i_manh))
    print("=" * 50)

    manh_error.append(np.mean(pred_i_manh != y_test))
```

```python
for k, accuracy in mean_scores_manh.items():
  print(f"k = {k}, Cross-validation mean accuracy: {accuracy[0] * 100}% with a standard deviation of {accuracy[1] * 100}")
```

### **4.3.1. Error v.s K-values plot (Manhattan Distance)**

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 51), manh_error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. k Value (Distance Metric: Manhattan)')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

## **4.4. Choosing Optimal k-Value**

```python
best_k_euc = max(mean_scores_euc, key=mean_scores_euc.get)
print(f"Best k: {best_k_euc} (Euclidean distance) with mean accuracy: {mean_scores_euc[best_k_euc][0] * 100}% and standard deviation {mean_scores_euc[best_k_euc][1]}")

best_k_manh = max(mean_scores_manh, key=mean_scores_manh.get)
print(f"Best k: {best_k_manh} (Manhattan distance) with mean accuracy: {mean_scores_manh[best_k_manh][0] * 100} % and standard deviation {mean_scores_euc[best_k_manh][1]}")

best_k_euc_2 = euc_error.index(min(euc_error)) + 1
print(f"Best k: {best_k_euc_2} (Euclidean distance) with mean error: {min(euc_error)}")

best_k_manh_2 = manh_error.index(min(manh_error)) + 1
print(f"Best k: {best_k_manh_2} (Manhattan distance) with mean error: {min(manh_error)}")
```

Since, there is a difference in the optimal value of k between the euclidean distance model and the manhattan distance model, we will choose the optimal k-values from both the models, train new models with these k-values in their respective distance metrics, and compare their results.

```python
# k-value
k = 14

# Initializing the model object with the necessary parameters
knn = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
# Fitting the model to the training data
knn.fit(X_train, y_train)
# Predicting the model using the testing data
pred = knn.predict(X_test)

# Printing various classification metrics to assess the performance of our model
print("Classification Report:")
print(classification_report(y_test, pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, pred))
print()
print(f"Accuracy score: {accuracy_score(y_test, pred)}")
print(f"Precision score: {precision_score(y_test, pred, average = 'macro')}")
print(f"Recall score: {recall_score(y_test, pred, average = 'macro')}")
print(f"F1 Score: {f1_score(y_test, pred, average = 'macro')}")
```

```python
# k-value
k = 9

# Initializing the model object with the necessary parameters
knn = KNeighborsClassifier(n_neighbors = k, metric = 'manhattan')
# Fitting the model to the training data
knn.fit(X_train, y_train)
# Predicting the model using the testing data
pred = knn.predict(X_test)

# Printing various classification metrics to assess the performance of our model
print("Classification Report:")
print(classification_report(y_test, pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, pred))
print()
print(f"Accuracy Score: {accuracy_score(y_test, pred)}")
print(f"Precision Score: {precision_score(y_test, pred, average = 'macro')}")
print(f"Recall Score: {recall_score(y_test, pred, average = 'macro')}")
print(f"F1 Score: {f1_score(y_test, pred, average = 'macro')}")
```

It can be seen from these classification metrics that the ***Manhattan*** distance k-NN model performs comparatively better than the Euclidean distance k-NN model. Thus, we will choose this as our desired k-NN model.

## **4.5. Training the Final Model**

```python
final_knn_model = KNeighborsClassifier(n_neighbors = 9, metric = 'manhattan')
final_knn_model.fit(X_train, y_train)
y_pred_knn = final_knn_model.predict(X_test)

# Printing various classification metrics to assess the performance of our model
print("Classification Report:")
print(classification_report(y_test, y_pred_knn))

print()

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))

print()

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_knn)}")
print(f"Precision Score: {precision_score(y_test, y_pred_knn, average = 'macro')}")
print(f"Recall Score: {recall_score(y_test, y_pred_knn, average = 'macro')}")
print(f"F1 Score: {f1_score(y_test, y_pred_knn, average = 'macro')}")
```

# **5. SVM Models**

## **5.1. Data Preparation**

### **5.1.1. Scaling the Features**

```python
X = df.iloc[:, :-1].values # Input Features
y = df.iloc[:, -1].values # Output column

# Generating a train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initializing a Standard scaler object
scaler = StandardScaler()
scaler.fit(X_train) # Training the scaler object on the training dataset only

# Scaling the feature values
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

## **5.2. Model(s) Training using Different Kernels and Evaluation**

```python
# List of all kernel types
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# Dictionary for storing the kernels as string keys and the corresponding average cross-val accuracy, standard deviations as list values
kernel_scores = dict()

for kernel in kernels:
  model = svm.SVC(kernel= kernel, class_weight='balanced', random_state=42)

  # Create a StratifiedKFold object
  cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

  # Apply cross-validation using StratifiedKFold
  scores = cross_val_score(model, X_train, y_train, cv=cv)
  kernel_scores[kernel] = [scores.mean(), scores.std()]

  model.fit(X_train, y_train)

  # Prediction and Validation
  y_pred = model.predict(X_test)

  print(f"Classification Report for {kernel} kernel:")
  print(classification_report(y_test, y_pred))

  print(f"Confusion Matrix for {kernel} kernel:")
  print(confusion_matrix(y_test, y_pred))
```

## **5.3. Choosing the Best Performing Kernel**

```python
for (kernel, score) in kernel_scores.items():
  print(f"Kernel: {kernel} => Cross-validation mean accuracy: {score[0] * 100}% with a standard deviation of {score[1] * 100}")
```

Overall, it appears that the "***rbf***" kernel seems to perform comparatively better (with a relatively low standard deviation) than the other kernels that were used. It not only has a higher accuracy % but also has better classification metrics and confusion matrix. Hence, we will consider the "rbf" kernel to be our primary kernel for the SVM model.

## **5.4. Model Training with Soft Margin and Hard Margin**

```python
# Training a soft margin SVM classifier with Regularization parameter C = 0.1
model_soft_margin = svm.SVC(C = 0.1, kernel= "rbf", class_weight='balanced', random_state = 42)
model_soft_margin.fit(X_train, y_train)

# Prediction and Validation
y_pred_soft = model_soft_margin.predict(X_test)

print(f"Classification Report:")
print(classification_report(y_test, y_pred_soft))

print(f"Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_soft))

print()

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_soft)}")
print(f"Precision Score: {precision_score(y_test, y_pred_soft, average = 'macro')}")
print(f"Recall Score: {recall_score(y_test, y_pred_soft, average = 'macro')}")
print(f"F1 Score: {f1_score(y_test, y_pred_soft, average = 'macro')}")

# Getting the support vectors
support_vectors_soft = model_soft_margin.support_vectors_

print("Support Vectors:", support_vectors_soft)
```

```python
# Training a hard margin SVM classifier with Regularization parameter C = 1e10
model_hard_margin = svm.SVC(C = 1e10, kernel= "rbf", class_weight='balanced', random_state = 42)
model_hard_margin.fit(X_train, y_train)

# Prediction and Validation
y_pred_hard = model_hard_margin.predict(X_test)

print(f"Classification Report:")
print(classification_report(y_test, y_pred_hard))

print(f"Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_hard))

print()

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_hard)}")
print(f"Precision Score: {precision_score(y_test, y_pred_hard, average = 'macro')}")
print(f"Recall Score: {recall_score(y_test, y_pred_hard, average = 'macro')}")
print(f"F1 Score: {f1_score(y_test, y_pred_hard, average = 'macro')}")

# Getting the support vectors
support_vectors_hard = model_hard_margin.support_vectors_

print("Support Vectors:", support_vectors_hard)
```

- It can be observed from the above provided metrics for the soft-margin and
hard-margin classifiers that the hard-margin classifier performs significantly better than the soft-margin classifier.

- All of the metrics of the hard-margin classifier are above 0.9, indicating a good performance.

- In contrast, all the metrics for the soft-margin classifier are below 0.8, indicating that the classifier is unable to create a good decision boundary.

## **6.1. Data Preparation**

### **6.1.1. Scaling the Features**

```python
X = df.iloc[:, :-1].values # Input Features
y = df.iloc[:, -1].values # Output column

# Generating a train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initializing a Standard scaler object
scaler = StandardScaler()
scaler.fit(X_train) # Training the scaler object on the training dataset only

# Scaling the feature values
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Creating a dictionary to store the cross-validation scores for different neural network models
cross_val_dict = dict()
```

## **6.2. Neural Network Model Architecture Selection**

### **6.2.1. Neural Network - 1**

#### Neural Network model with **no Dropout layer(s)** and **2 hidden layers**.

---

#### **Neural Network Architecture used:**

- Hidden Layer - 1: 10 neurons
- Hidden Layer - 2: 5 neurons
- Output Layer: 3 neurons

#### **Activation Functions used:**
- Hidden Layer(s): ReLU Function
- Output Layer: Softmax Function

```python
# Function to create a neural network model
def create_model_1():
  tf.random.set_seed(42)

  model_1 = tf.keras.Sequential()

  model_1.add(tf.keras.layers.Dense(10, activation = 'relu', input_shape = (X_train.shape[1],)))
  model_1.add(tf.keras.layers.Dense(5, activation = 'relu'))
  model_1.add(tf.keras.layers.Dense(3, activation = 'softmax'))

  model_1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  return model_1
```

```python
# Creating a KerasClassifier object for cross-validation using Stratified K-fold
model_1 = KerasClassifier(build_fn = create_model_1, epochs=100, batch_size=100, verbose=0)

stratified_kfold = StratifiedKFold(n_splits = 10, shuffle=True, random_state=42)
cross_val_scores = cross_val_score(model_1, X_train, y_train, cv=stratified_kfold)
```

```python
cross_val_dict['model_1'] = cross_val_scores
```

### **6.2.2. Neural Network - 2**

#### Neural Network model with **Dropout layer(s)** included and **2 hidden layers**.

---

#### **Neural Network Architecture used:**

- Hidden Layer - 1: 20 neurons
- Dropout Layer - 1: 25% neurons to dropout
- Hidden Layer - 2: 15 neurons
- Dropout Layer - 2: 25% neurons to dropout
- Output Layer: 3 neurons

#### **Activation Functions used:**
- Hidden Layer(s): ReLU Function
- Output Layer: Softmax Function

```python
# Function to create a neural network model
def create_model_2():
  tf.random.set_seed(42)

  model_2 = tf.keras.Sequential()

  model_2.add(tf.keras.layers.Dense(20, activation = 'relu', input_shape = (X_train.shape[1],)))
  model_2.add(tf.keras.layers.Dropout(0.25))
  model_2.add(tf.keras.layers.Dense(15, activation = 'relu'))
  model_2.add(tf.keras.layers.Dropout(0.25))
  model_2.add(tf.keras.layers.Dense(3, activation = 'softmax'))

  model_2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  return model_2
```

```python
# Creating a KerasClassifier object for cross-validation using Stratified K-fold
model_2 = KerasClassifier(build_fn = create_model_2, epochs=100, batch_size=100, verbose=0)

stratified_kfold = StratifiedKFold(n_splits = 10, shuffle=True, random_state=42)
cross_val_scores_2 = cross_val_score(model_2, X_train, y_train, cv=stratified_kfold)
```

```python
cross_val_dict['model_2'] = cross_val_scores_2
```

### **6.2.3. Neural Network - 3**

#### Neural Network model with **Dropout layer(s)** included and **3 hidden layers**.

---

#### **Neural Network Architecture used:**

- Hidden Layer - 1: 20 neurons
- Dropout Layer - 1: 20% neurons to dropout
- Hidden Layer - 2: 15 neurons
- Dropout Layer - 2: 20% neurons to dropout
- Hidden Layer - 3: 10 neurons
- Dropout Layer - 3: 20% neurons to dropout
- Output Layer: 3 neurons

#### **Activation Functions used:**
- Hidden Layer(s): ReLU Function

- We will now train our model using the optimal range of α values and batch size values we found through our observations above.

- We will train the model 10 times and then average the performance to get a more stable number for the model's performance.

```python
number_of_times_trained = 10 # Number of times to train the same model
accuracy_list = list() # Creating a list to store the testing accuracy for the model in each training round

for training_number in range(number_of_times_trained):
  first_model = create_model_3()
  first_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  history = first_model.fit(X_train, y_train, epochs=100, batch_size=64, verbose = 0, validation_split=0.2)

  # Evaluating the model on the testing dataset
  test_loss, test_accuracy = first_model.evaluate(X_test, y_test)
  accuracy_list.append(test_accuracy)

  # Print the testing accuracy
  print(f"Testing Accuracy for Training Run {training_number + 1}: {test_accuracy * 100:.2f}%")

average_accuracy = np.mean(accuracy_list)
print(f"Average Testing Accuracy: {average_accuracy * 100:.2f}%")
```

As can be seen from the above results, the model has a 90%+ accuracy (more than 92%). This is close to the cross-validation accuracy we had observed before and is by itself a pretty decent accuracy percentage.
